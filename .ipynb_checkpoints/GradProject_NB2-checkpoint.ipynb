```python
# Returns the pixel size of the image
def ft0(image):
    return np.array(image).shape

# Returns the average of the red-channel pictures for the images
def ft1(image):
    pil = Image.fromarray(image)
    return np.mean(pil.getdata(band=0))
# Source: https://stackoverflow.com/questions/51325224/python-pil-image-split-to-rgb (2)

#Returns the aspect ratio of the image
def ft2(image):
    size = np.array(image).shape
    height = size[0]
    width = size[1]
    return float(width/height)
# Source: https://en.wikipedia.org/wiki/Aspect_ratio_(image) (1)
    
# Returns the average of the green-channel pictures for the images
def ft3(image):
    pil = Image.fromarray(image)
    return np.mean(image.getdata(band=1))
# Source: https://stackoverflow.com/questions/51325224/python-pil-image-split-to-rgb (2)

# Returns the average of the blue-channel pictures for the images
def ft4(image):
    pil = Image.fromarray(image)
    return np.mean(pil.getdata(band=2))
# Source: https://stackoverflow.com/questions/51325224/python-pil-image-split-to-rgb (2)

# Returns a matrix of descriptors after ORB corner detection
# Each descriptor is computed from a keypoint. A keypoint is a circular image region with an orientation computed by the algorithm.
def ft5(image):
    # Convert image into HSV
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2HSV) #HSV format
    # initiate STAR detector
    orb = cv2.ORB_create()
    # find the keypoints and descriptors with ORB
    keypoints, descriptors = orb.detectAndCompute(gray, None)
    return descriptors
# Source: https://stackoverflow.com/questions/14134892/convert-image-from-pil-to-opencv-format (3)
# Source: https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_feature2d/py_orb/py_orb.html#orb (4)
# Source: http://www.vlfeat.org/api/sift.html

# Returns a matrix of descriptors after ORB corner detection
# Each descriptor is computed from a keypoint. A keypoint is a circular image region with an orientation computed by the algorithm.
def ft6(image):
    # convert image into grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) #GRAYSCALE format
    # initiate STAR detector
    orb = cv2.ORB_create()
    # find the keypoints and descriptors with ORB
    keypoints, descriptors = orb.detectAndCompute(gray, None)
    return descriptors
# Source: https://stackoverflow.com/questions/14134892/convert-image-from-pil-to-opencv-format (3)
# Source: https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_feature2d/py_orb/py_orb.html#orb (4)
# Source: http://www.vlfeat.org/api/sift.html

# Returns value of the darkest pixel in grayscale image. Remember that 0 = white, 1 = black
def ft7(image):
    pil = Image.fromarray(image, mode='L') # Fetch image in grayscale
    extreme = pil.getextrema() # Fetch min and max pixels
    return extreme[1]

# Return most common (R,G,B) color in the photo
def ft8(image):
    k = 150 # Let k be any constant
    pil = Image.fromarray(image)
    pil = pil.resize((k, k)) # For speed
    colors = pil.getcolors(k*k) 
    return colors #max(colors)[0]
    
# Return mean of Haralick algorithm for texture feature
def ft9(image):
    # convert image into grayscale
    img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # calculate haralick texture features for 4 types of adjacency
    textures = mahotas.features.haralick(img)
    # take the mean of it and return it
    haralick = textures.mean(axis=0)
    return haralick
# Source: http://murphylab.web.cmu.edu/publications/boland/boland_node26.html

# Return color histogram for HSV color space
def ft10(image):
    # convert the image to HSV color space
    img = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    # compute the color histogram
    hist  = cv2.calcHist([img], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    # normalize the histogram
    cv2.normalize(hist, hist)
    # return the histogram
    return hist.flatten()

# Return color histogram for Grayscale
def ft10(image):
    # convert the image to HSV color space
    img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # compute the color histogram
    hist  = cv2.calcHist([img], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    # normalize the histogram
    cv2.normalize(hist, hist)
    # return the histogram
    return hist.flatten()

# Return Hu Moments. Here we convert to threshold image
def ft11(image):
    img_cv = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # convert image to a binary image using Threshold img conversion
    _, img_thresh = cv2.threshold(img_cv, 128, 255, cv2.THRESH_BINARY)
    hu = cv2.HuMoments(cv2.moments(img_thresh)).flatten()
    # transform hu moments by using log scale formula
    log_hu = [-1 * copysign(1.0, h) * log10(abs(h)) for h in hu]
    return log_hu

def ft11_revised(image):
    img_cv = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # convert image to a binary image using Threshold img conversion
    #_, img_thresh = cv2.threshold(img_cv, 128, 255, cv2.THRESH_BINARY)
    # transform hu moments by using log scale formula
    hu = cv2.HuMoments(cv2.moments(img_cv)).flatten()
    log_hu = [-1 * copysign(1.0, h) * log10(abs(h)) for h in hu]
    return log_hu

# Why use Hu Moments? Check out this resource https://www.learnopencv.com/shape-matching-using-hu-moments-c-python/
# Hu moments are image moments that are invariant to translation, scale, and rotation
```
