```python
# Returns the pixel size of the image
def ft0(image):
    return image.size

# Returns the average of the red-channel pictures for the images
def ft1(image):
    return np.mean(image.getdata(band=0))
# Source: https://stackoverflow.com/questions/51325224/python-pil-image-split-to-rgb (2)


#Returns the aspect ratio of the image
def ft2(image):
    return float(image.width)/(image.height)
# Source: https://en.wikipedia.org/wiki/Aspect_ratio_(image) (1)
    

# Returns the average of the green-channel pictures for the images
def ft3(image):
    return np.mean(image.getdata(band=1))
# Source: https://stackoverflow.com/questions/51325224/python-pil-image-split-to-rgb (2)

# Returns the average of the blue-channel pictures for the images
def ft4(image):
    return np.mean(image.getdata(band=2))
# Source: https://stackoverflow.com/questions/51325224/python-pil-image-split-to-rgb (2)

def ft5(image):
    return np.mean(image.getdata(band=2))

# Returns a matrix of descriptors after ORB corner detection
# Each descriptor is computed from a keypoint. A keypoint is a circular image region with an orientation computed by the algorithm.
def ft6(image):
    # Convert image into grayscale
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2GRAY)
    # Initiate STAR detector
    orb = cv2.ORB_create()
    # Find the keypoints and descriptors with ORB
    keypoints, descriptors = orb.detectAndCompute(gray, None)
    return descriptors
# Source: https://stackoverflow.com/questions/14134892/convert-image-from-pil-to-opencv-format (3)
# Source: https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_feature2d/py_orb/py_orb.html#orb (4)
# Source: http://www.vlfeat.org/api/sift.html

# Returns value of the darkest pixel in grayscale image. Remember that 0 = white, 1 = black
def ft7(image):
    extreme = image.convert("L").getextrema() # Fetch min and max pixels
    return extreme[1]

# Return most common (R,G,B) color in the photo
def ft8(image):
    k = 150 # Let k be any constant
    img = image.resize((k, k)) # For speed
    colors = img.getcolors(k*k) 
    return max(colors)[1]
    
# Return mean of Haralick algorithm for texture feature
def ft9(image):
    # Convert image into grayscale
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2GRAY)
    # Calculate haralick texture features for 4 types of adjacency
    textures = mahotas.features.haralick(gray)
    # take the mean of it and return it
    ht_mean = textures.mean(axis=0)
    return ht_mean
# Source: http://murphylab.web.cmu.edu/publications/boland/boland_node26.html
```
